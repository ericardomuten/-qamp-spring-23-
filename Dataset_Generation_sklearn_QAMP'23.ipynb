```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

# Generate a dataset with a circle pattern
X_circle, y_circle = make_circles(n_samples=1000, noise=0.1, factor=0.5)

# Plot the circle pattern
plt.scatter(X_circle[:, 0], X_circle[:, 1], c=y_circle)
plt.title('Circle Pattern')
plt.show()
```

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# Generate a dataset with a moons pattern
X_moons, y_moons = make_moons(n_samples=1000, noise=0.1)

# Plot the moons pattern
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons)
plt.title('Moons Pattern')
plt.show()
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate a dataset with a blobs pattern
X_blobs, y_blobs = make_blobs(n_samples=1000, centers=3)

# Plot the blobs pattern
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_blobs)
plt.title('Blobs Pattern')
plt.show()
```

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

# Generate a dataset with concentric circles
X_concentric, y_concentric = make_circles(n_samples=1000, noise=0.05, factor=0.3)

# Plot the concentric circles pattern
plt.scatter(X_concentric[:, 0], X_concentric[:, 1], c=y_concentric)
plt.title('Concentric Circles Pattern')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset with a diamond shape
np.random.seed(0)
X_diamond = np.random.rand(1000, 2)
X_diamond[:, 1] *= 4
X_diamond[:, 0] += np.abs(X_diamond[:, 1] - 2) * 0.5
y_diamond = np.logical_and(X_diamond[:, 0] > 1, X_diamond[:, 0] < 3) & (X_diamond[:, 1] > 1) & (X_diamond[:, 1] < 3)

# Plot the diamond pattern
plt.scatter(X_diamond[:, 0], X_diamond[:, 1], c=y_diamond)
plt.title('Diamond Shape Pattern')
plt.show()
```
